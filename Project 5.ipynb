```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the dataset
data = pd.read_csv("D:\studies girl\certificate\epa-sea-level.csv")
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Year'], data['CSIRO Adjusted Sea Level'], color='blue', label='Data')
```

```python
# Line of best fit for all data
slope, intercept, r_value, p_value, std_err = linregress(data['Year'], data['CSIRO Adjusted Sea Level'])
plt.plot(data['Year'], slope * data['Year'] + intercept, color='red', label='Fit Line')

# Extend the line to 2050
years_extended = pd.Series(range(data['Year'].min(), 2051))
plt.plot(years_extended, slope * years_extended + intercept, color='red', linestyle='--')
```

```python
# Filter data from the year 2000
data_recent = data[data['Year'] >= 2000]

# Line of best fit for recent data
slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(data_recent['Year'], data_recent['CSIRO Adjusted Sea Level'])
plt.plot(data_recent['Year'], slope_recent * data_recent['Year'] + intercept_recent, color='green', label='Fit Line (2000 onwards)')

# Extend the line to 2050
plt.plot(years_extended, slope_recent * years_extended + intercept_recent, color='green', linestyle='--')
```

```python
# Customize plot
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()
```

```python
# Save the plot
plt.savefig('sea_level_plot.png')

# Show the plot
plt.show()
```
